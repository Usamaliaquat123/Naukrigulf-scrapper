```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver as wd
import pandas as pd
```

```python
url="https://www.naukri.com/mnc-jobs?src=discovery_trendingWdgt_homepage_srch"
```

```python
header={'brave':'huzaifa'}
```

```python
web=wd.Chrome()
```

```python
web.get(url)
```

```python
content=bs(web.page_source,'html.parser')
```

```python
content
```

```python
response=requests.get(url,headers=header)
```

```python
response
```

```python
job_titles=[]
for i in content.find_all('div',class_="row1"):
    job_titles.append(i.text)
job_titles
```

```python
company_name=[]
for i in content.find_all('span',class_="comp-dtls-wrap"):
    company_name.append(i.text)
company_name
```

```python
locations=[]
for i in content.find_all('span','loc-wrap ver-line'):
    locations.append(i.text)
locations
```

```python
job_description=[]
for i in content.find_all('div',class_='row5'):
    job_description.append(i.text)
len(job_description)
```

```python
experience=[]
for i in content.find_all('span',class_='exp-wrap'):
    experience.append(i.text)
experience
```

```python
assert len(job_titles)==len(company_name)==len(job_description)==len(experience)==len(locations)
```

```python
data={'Role':job_titles,'Company_name':company_name,'Description':job_description,'Experience':experience,'Location':locations}
csv='naukri.csv'
df=pd.DataFrame(data)
df.to_csv(csv,index=False,encoding='utf-8')
print('created csv file sucessfully')
```
